```python
merged_df = pd.merge(offer_completions_df,signups_df, on=['user_id', 'app_id'])
merged_df
```

```python
merged_df = pd.merge(merged_df, rewards_df, on='reward_id')
merged_df
```

```python
merged_df.isnull().sum()
```

```python

# Calculate the metrics
arpu = merged_df['total_revenue_in_paise'].sum() / merged_df['user_id'].nunique()
avg_lifespan = (merged_df['last_login_date'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d'))- merged_df['date'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d'))).mean().days
clv = arpu * avg_lifespan
total_users = merged_df['user_id'].nunique()
ltv = clv * total_users/100

# Print the results
print(f'ARPU: {arpu}')
print(f'Average Lifespan: {avg_lifespan} days')
print(f'CLV: {clv}')
print(f'Total Users: {total_users}')
print(f'LTV: {ltv} rupees')
```

```python
import pandas as pd
import statsmodels.api as sm
import numpy as np

# read in the October data
data = pd.read_csv('/content/Mydrive/MyDrive/Greedygame/Copy of Q3_Sikka data.csv')
oct_data=data[(data['Date'] >= '2022-10-01') & (data['Date'] < '2022-11-01')]
nov_data=data[(data['Date'] >= '2022-11-01') & (data['Date'] < '2022-11-16')]
```

```python
oct_data
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(12,12))
sns.heatmap(oct_data.corr(), annot=True, linewidths=.5 , ax=ax,cmap="coolwarm")
plt.show()
```

```python
nov_data
```

```python

# create a new column for the log of daily active users
oct_data['log_DAU'] = np.log(oct_data['DAU'])

# create a linear regression model
X = oct_data[['Signups', 'log_DAU']]
y = oct_data['Referrals']
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()

# print the model summary
print(model.summary())
```

```python
# read in the November data

# create a new column for the log of daily active users
nov_data['log_DAU'] = np.log(nov_data['DAU'])

# use the model to predict the number of referrals
nov_data['predicted_referrals'] = model.predict(sm.add_constant(nov_data[['Signups', 'log_DAU']]))

# print the predicted number of referrals for each day
print(nov_data['predicted_referrals'])
```

```python
# merge the offer_data_df and user_signup_df tables on the user_id column
merged_offer_data = pd.merge(offer_data_df, signups_df[['user_id', 'app_id']], on='user_id')
```

```python
# count the number of unique user_id in the merged_offer_data table for each app
offer_initiation_counts = merged_offer_data.groupby('app_id')['user_id'].nunique()
offer_initiation_counts
```

```python
# count the number of unique user_id in the offer_completion_df table for each app
offer_completion_counts = merged_offer_data[merged_offer_data['status'] == 'COMPLETED'].groupby('app_id')['user_id'].nunique()
offer_completion_counts
```

```python
# sum up the total_payout_in_paise column in the rewards_df table for each app
rewards_earned = merged_df.groupby('app_id')['total_payout_in_paise'].sum() / 100 # divide by 100 to convert from paise to rupees
rewards_earned
```

```python
# sum up the total_revenue_in_paise column in the rewards_df table for each app
revenue_generated = merged_df.groupby('app_id')['total_revenue_in_paise'].sum() / 100 # divide by 100 to convert from paise to rupees
revenue_generated
```

```python
import matplotlib.pyplot as plt

# create a bar chart to compare the performance of both the apps
fig, ax = plt.subplots(figsize=(8, 6))
offer_initiation_counts.plot(kind='bar', ax=ax, color='orange', alpha=0.7, label='Offer Initiation')
offer_completion_counts.plot(kind='bar', ax=ax, color='green', alpha=0.7, label='Offer Completion')
rewards_earned.plot(kind='bar', ax=ax, color='blue', alpha=0.7, label='Rewards Earned')
revenue_generated.plot(kind='bar', ax=ax, color='red', alpha=0.7, label='Revenue Generated')
ax.set_title('Performance Comparison of Sikka and Sikka Pro')
ax.set_xlabel('App')
ax.set_ylabel('Count / Rupees')
ax.legend()
plt.show()
```
