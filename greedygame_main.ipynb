```python
import pandas as pd
import statsmodels.api as sm
import numpy as np

# read in the October data
data = pd.read_csv('/content/Mydrive/MyDrive/Greedygame/Copy of Q3_Sikka data.csv')
oct_data=data[(data['Date'] >= '2022-10-01') & (data['Date'] < '2022-11-01')]
nov_data=data[(data['Date'] >= '2022-11-01') & (data['Date'] < '2022-11-16')]
```

```python
oct_data
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(12,12))
sns.heatmap(oct_data.corr(), annot=True, linewidths=.5 , ax=ax,cmap="coolwarm")
plt.show()
```

```python
nov_data
```

```python

# create a new column for the log of daily active users
oct_data['log_DAU'] = np.log(oct_data['DAU'])

# create a linear regression model
X = oct_data[['Signups', 'log_DAU']]
y = oct_data['Referrals']
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()

# print the model summary
print(model.summary())
```

```python
# read in the November data

# create a new column for the log of daily active users
nov_data['log_DAU'] = np.log(nov_data['DAU'])

# use the model to predict the number of referrals
nov_data['predicted_referrals'] = model.predict(sm.add_constant(nov_data[['Signups', 'log_DAU']]))

# print the predicted number of referrals for each day
print(nov_data['predicted_referrals'])
```

```python
# sum up the total_payout_in_paise column in the rewards_df table for each app
rewards_earned = merged_df.groupby('app_id')['total_payout_in_paise'].sum() / 100 # divide by 100 to convert from paise to rupees
rewards_earned
```

```python
# Function to find the data point that is more than 3 times the standard deviation
def find_anomalies(data):
    #define a list to accumlate anomalies
    anomalies = []
    
    # Set upper and lower limit to 3 standard deviation
    random_data_std = np.std(data)
    random_data_mean = np.mean(data)
    anomaly_cut_off = random_data_std * 3
    
    lower_limit  = random_data_mean - anomaly_cut_off 
    upper_limit = random_data_mean + anomaly_cut_off
    print(lower_limit)
    # Generate outliers
    for outlier in data:
        if outlier > upper_limit or outlier < lower_limit:
            anomalies.append(outlier)
    return anomalies
```
